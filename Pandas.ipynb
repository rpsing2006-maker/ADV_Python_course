```python
from textblob import TextBlob
```

```python
def textblob_sentiment_analysis(review):
  sentiment = TextBlob(review).sentiment
  if sentiment.polarity > 0.1:
    return 'positive'
  elif sentiment.polarity < -0.1:
    return 'negative'
  else:
    return 'neutral'
```

```python
df['Sentiment'] = df['Review'].apply(textblob_sentiment_analysis)
```

```python
df.sample(5)
```

```python
sentiment_distribution = df['Sentiment'].value_counts()
sentiment_distribution
```

```python
plt.figure(figsize = (9,5))
sns.countplot(data = df ,x = 'Sentiment')
plt.title("distribution of rating")
plt.xlabel('rating')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize = (9,5))
sns.countplot(data = df ,x = 'Rating',hue = 'Sentiment')
plt.xlabel('rating')
plt.ylabel('count')
plt.legend(title = 'sentiment')
plt.show()
```

```python
df = pd.read_csv("/content/supply_chain - supply_chain.csv")
df.head()
```

```python
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
pio.templates.defalut = "plotly_white"
```

```python
df.describe()
```

```python
!pip install statsmodels
```

```python
fig = px.scatter(df , x="Price",
                 y = "Revenue generated",
                 color = "Product type",
                 hover_data= ["Number of products sold"],
                 trendline = 'ols')
fig.show()
```

```python
sales_data = df.groupby('Product type')['Number of products sold'].sum().reset_index()
sales_data

pie_chart = px.pie(sales_data, values='Number of products sold', names='Product type', title='Sales by Product Type', hover_data = ['Number of products sold'],hole = 0.5 , color_discrete_sequence=px.colors.qualitative.Pastel)
pie_chart.update_traces(textposition='inside', textinfo='percent+label')
pie_chart.show()
```

```python
total_revenue = df.groupby('Shipping carriers')['Revenue generated'].sum().reset_index()
fig = go.Figure()
fig.add_trace(go.Bar(x = total_revenue['Shipping carriers'],
                    y = total_revenue['Revenue generated'],
                    name = 'Total Revenue'))
fig.update_layout(title = 'Total Revenue by Shipping Carriers',
                xaxis_title = 'Shipping Carriers',
                yaxis_title = 'Total Revenue')
fig.show()
```

```python
#now lets have look on avg lead time and avg manufacturing cost for all the products in the company
avg_lead_time = df.groupby('Product type')['Lead time'].mean().reset_index()
# avg_lead_time
avg_manufacturing_cost = df.groupby('Product type')['Manufacturing costs'].mean().reset_index()
result = pd.merge(avg_lead_time,avg_manufacturing_cost,on = 'Product type')
result.rename(columns={'Lead time':'Avg Lead Time','Manufacturing costs':'Avg Manufacturing Cost'},inplace = True)
print(result)
```
